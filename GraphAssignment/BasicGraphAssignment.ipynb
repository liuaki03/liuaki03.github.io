### Import Modules

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

### Opiods in Virginia

```python
# Import Data
opiod_df = pd.read_csv("OpidsVA.csv")
```

```python
opiod_df.head()
```

#### Q1: Do opioid overdoes tend to be associated with less affluent areas - that is, areas where families have lower incomes?

```python
# I used a scatterplot to show correlation since this question asked about the 
# relationship between opioid overdoses and household income.

fig,ax = plt.subplots()
ax.scatter(opiod_df['MedianHouseholdIncome'], opiod_df['FPOO-Rate'], alpha=0.5, c = 'b')
fig.suptitle('Opioid Overdose vs. Median Household Income')
ax.xaxis.set_label_text('Median Household Income') 
ax.yaxis.set_label_text('Opioid Overdoses per 100,000 County Residents') 
fig.set_size_inches(8,6)
plt.show()
```

#### Q2: Some people who start with opioid addictions are reported to transition to heroin use. What is the relationship in Virginia counties between opioid overdoses and heroin overdoses?

```python
# I used a scatterplot to show correlation since this question asked about the relationship 
# between opioid overdoses and heroin overdoses.

fig,ax = plt.subplots()
ax.scatter(opiod_df['FFHO-Rate'], opiod_df['FPOO-Rate'], alpha=0.5, c = 'b')
fig.suptitle('Opioid Overdose vs. Heroin Overdoses')
ax.xaxis.set_label_text('Heroin Overdoses per 100,000 County Residents') 
ax.yaxis.set_label_text('Opioid Overdoses per 100,000 County Residents') 
fig.set_size_inches(8,6)
plt.show()
```

### United States Presidents

```python
# Import Data
presidents_df = pd.read_csv("presidents.csv")
```

```python
presidents_df.head()
```

#### Q1: Which states are associated with the greatest number of United States presidents in terms of the presidents' birthplaces?

```python
# I used a bar chart to show the frequency distribution of US presidential births in each state
# since the question asked which states had the greatest number of presidential births.

# Step 1: Create frequency df
presidents_freq = pd.value_counts(presidents_df["State"]).reset_index().rename(columns={"index":"State",
                                                                                        "State":"Count"})
# Step 2: Plot graph
fig,ax = plt.subplots()
ax.bar(presidents_freq["State"],presidents_freq["Count"])
fig.suptitle('Number of United States Presidents\' Births in Each State')
ax.xaxis.set_label_text('State') 
ax.yaxis.set_label_text('Number of Presidential Births') 
fig.set_size_inches(12,8)
plt.show()
```

### United States Vehicle Sales

```python
# Import Data
vehicles_df = pd.read_csv("TOTALNSA.csv")
```

```python
vehicles_df.head()
```

#### Q1: How have vehicle sales in the United States varied over time?

```python
# I used a line graph to show changes over time since the question asked about variation over time 
# and the file contains time-series vehicle data.

fig,ax = plt.subplots()
ax.plot(vehicles_df['DATE'], vehicles_df['TOTALNSA'])
fig.suptitle('Vehicle Sales in the United States Over Time')
ax.xaxis.set_label_text('Monthly Date') 
ax.yaxis.set_label_text('Total Sales') 
fig.set_size_inches(14,8)
plt.show()
```
